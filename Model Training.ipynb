```python
import pandas as pd
```

```python
df = pd.read_csv(r'day.csv')
df.iloc[147]
```

```python

df=df.drop(labels=['instant','dteday','casual','registered','atemp'],axis=1)
```

```python
df['season']=df['season'].astype('category')
df['weathersit']=df['weathersit'].astype('category')
df['mnth']=df['mnth'].astype('category')
df['weekday']=df['weekday'].astype('category')
df = pd.get_dummies(df,dtype='int')
df.info()
```

```python

X = df.drop(labels=['cnt'],axis=1)
Y = df[['cnt']]
```

```python

Y
```

```python
categorical_cols = X.select_dtypes(include='object').columns
numerical_cols = X.select_dtypes(exclude='object').columns
```

```python
numerical_cols
```

```python
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import MinMaxScaler 
from sklearn.preprocessing import OrdinalEncoder 
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
!pip install joblib
```

```python
import joblib
num_pipeline=Pipeline(
    steps=[
    ('imputer',SimpleImputer(strategy='median')),
    ('scaler',MinMaxScaler())

    ]

)
cat_pipeline=Pipeline(
    steps=[
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('ordinalencoder',OrdinalEncoder(categories=[])),
    ('scaler',MinMaxScaler())
    ]

)

preprocessor=ColumnTransformer([
('num_pipeline',num_pipeline,numerical_cols),
('cat_pipeline',cat_pipeline,categorical_cols)
])
```

```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=30)
pickle.dump(X_train.iloc[0],open('X_train_new.pkl','wb'))
```

```python
joblib.dump(preprocessor,"pre_processor.x")
```

```python
X_train=pd.DataFrame(preprocessor.fit(X_train),columns=preprocessor.get_feature_names_out())
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe1 = RFE(lm)
rfe1 = rfe1.fit(X_train, y_train)
```

```python
import statsmodels.api as sm  

 

# Subset the features selected by rfe1
col1 = X_train.columns[rfe1.support_]

 

# Subsetting training data for 5t selected columns
X_train_rfe1 = X_train[col1]

 

# Add a constant to the model
X_train_rfe1 = sm.add_constant(X_train_rfe1)
X_train_rfe1.head()
```

```python
y_train
```

```python
y_train.reset_index(drop=True,inplace=True)
```

```python
lm1 = sm.OLS(y_train, X_train_rfe1).fit()   

print(lm1.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()

vif['Features'] = X_train_rfe1.columns

vif['VIF'] = [variance_inflation_factor(X_train_rfe1.values, i) for i in range(X_train_rfe1.shape[1])]

vif['VIF'] = round(vif['VIF'], 2)

vif = vif.sort_values(by = "VIF", ascending = False)

vif
```
