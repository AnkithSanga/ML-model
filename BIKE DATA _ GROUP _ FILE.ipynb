```python
df['season']=df['season'].astype('category')
df['weathersit']=df['weathersit'].astype('category')
df['mnth']=df['mnth'].astype('category')
df['weekday']=df['weekday'].astype('category')

df.info()
```

```python
df = pd.get_dummies(df, drop_first=True,dtype='int')
df.info()
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, train_size = 0.70, test_size = 0.30, random_state = 100)
```

```python
df_test
```

```python
def f():
    from sklearn.preprocessing import MinMaxScaler
    s = MinMaxScaler()
    num_vars = ['temp', 'hum', 'windspeed','cnt']

    df_train[num_vars] = s.fit_transform(df_train[num_vars])
```

```python
f()
df_train
```

```python
df_train.describe()
```

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe1 = RFE(lm)
rfe1 = rfe1.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe1.support_,rfe1.ranking_))
```

```python

import statsmodels.api as sm  

 

# Subset the features selected by rfe1
col1 = X_train.columns[rfe1.support_]

 

# Subsetting training data for 5t selected columns
X_train_rfe1 = X_train[col1]

 

# Add a constant to the model
X_train_rfe1 = sm.add_constant(X_train_rfe1)
X_train_rfe1.head()
```
