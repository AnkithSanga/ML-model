```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the data
data = pd.read_csv('day.csv',parse_dates=True) 

# Preprocess the data


```

```python
data=data.drop(['instant'],axis=1)
data.head()
```

```python
# Normalize the features
scaler = StandardScaler()
scaler.fit(data['cnt'].values.reshape(-1,1))
data['cnt'] = scaler.transform(data['cnt'].values.reshape(-1, 1))
```

```python
data.iloc[:,:-1]
```

```python
# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,:-1], data.iloc[:,-1], test_size=0.1, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#Reshaping the data to adjust to the tensor
# Split the data into train and test sets

X_train_reshape = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_reshape = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))
#Reshaping the data to adjust to the tensor
# Split the data into train and test sets

y_train_reshape = y_train.reshape((y_train.shape[0], 1, y_train.shape[1]))
y_test_reshape = y_test.reshape((y_test.shape[0], 1, y_test.shape[1]))
```

```python
X_train_reshape.shape
```

```python
rnn = Sequential()

rnn.add(LSTM(128, input_shape=(X_train_reshape.shape[1], X_train_reshape.shape[2]))) # Pass 2nd and 3rd dimensions from the input matrix
rnn.add(Dense(units=1))
```

```python
rnn.summary()
```

```python
rnn.compile(optimizer='SGD', loss='mse')
results = rnn.fit(X_train_reshape, y_train, validation_data=(X_test_reshape, y_test), epochs=20)
```

```python
X_train.shape
# y_train.shape
```

```python
rnn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
rnn.fit(X_train, y_train, epochs=10, batch_size=32)

# Evaluate the model
loss, accuracy = rnn.evaluate(X_test, y_test)
print('Test Loss:', loss)
print('Test Accuracy:', accuracy)

# Make predictions
predictions = rnn.predict(X_test)
```
